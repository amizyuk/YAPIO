```python
import numpy as np
```

```python
def _season_from_months(months):
    """Compute season (DJF, MAM, JJA, SON) from month ordinal"""
    seasons = np.array(["DJF", "MAM", "JJA", "SON", "nan"])
    idx = (months.values // 3) % 4
    idx[np.isnan(idx)] = 4
    return seasons[idx.astype(int)]
```

```python
cvdf["season"] = _season_from_months(cvdf["month"])
```

### Среднеквадратическая ошибка

$$
RMSE = \sqrt{{ 1 \over N} \sum \limits _{i} ^ {N} (x - x_{obs})^2}
$$

```python
from numpy import sqrt
def rmse(x):
    return sqrt((x**2).mean())
```

```python
sel[["dthetao","dso"]].agg(("mean",rmse,"min","max","count"))
```
