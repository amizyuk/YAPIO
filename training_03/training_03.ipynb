```python
from matplotlib.colors import TwoSlopeNorm
```

```python
plt.figure(figsize=(10,3))
plt.pcolormesh(time,z,ptem,cmap="RdBu_r",norm=TwoSlopeNorm(vmin=7.,vcenter=10.,vmax=30.))
plt.gca().invert_yaxis()
plt.colorbar(pad=0.015,shrink=0.7,label="T,\u00b0C")
```

```python
plt.figure(figsize=(10,3))
plt.pcolormesh(time,z,ke)
plt.gca().invert_yaxis()
plt.colorbar(pad=0.015,shrink=0.7,label="KE,m\u00b2/s\u00b2")
```

```python
from matplotlib.colors import LogNorm
```

```python
plt.figure(figsize=(10,3))
plt.pcolormesh(time,z,ke,norm=LogNorm(vmin=1.e-4,vmax=1.e-1))
plt.gca().invert_yaxis()
plt.colorbar(pad=0.015,shrink=0.7,label="KE,m\u00b2/s\u00b2")
```

```python
plt.figure(figsize=(10,3))
plt.pcolormesh(time,z,ke,norm=LogNorm(vmin=1.e-4,vmax=1.e-1),cmap="gist_stern")
plt.gca().invert_yaxis()
plt.colorbar(pad=0.015,shrink=0.7,label="KE,m\u00b2/s\u00b2")
```
